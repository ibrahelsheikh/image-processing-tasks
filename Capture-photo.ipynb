```python
import cv2
from matplotlib import pyplot as plt
```

```python
# capture photo from webcam no 0
webcam = cv2.VideoCapture(0)

check, frame = webcam.read()
print(check)

# showing frame value 
print(frame)
cv2.imwrite(filename='saved_img.jpg', img=frame)
print ("Image saved!")
```
